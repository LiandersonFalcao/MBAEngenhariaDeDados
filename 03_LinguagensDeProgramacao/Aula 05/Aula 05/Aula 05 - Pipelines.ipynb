# DVC - Data Version Control

### 📦 O que é DVC?
DVC (Data Version Control) é uma ferramenta de versionamento de dados e pipelines para projetos de ciência e engenharia de dados. Ele traz para os dados e para os processos de transformação (ETL, ML, etc.) o que o Git já faz para o código:

| Recurso       | Git                           | DVC                        |
| ------------- | ----------------------------- | -------------------------- |
| Código        | `*.py`, `*.ipynb`             | Versionado por Git         |
| Dados grandes | `*.csv`, `*.parquet`, imagens | Versionados por DVC        |
| Pipelines     | `Makefile`, `scripts.sh`      | Versionados com `dvc.yaml` |


### 🔄 Como o DVC executa um pipeline?
O DVC organiza cada passo do pipeline em stages, registrados no arquivo dvc.yaml. Cada stage contém:

 - **cmd**: o comando Python ou script a executar
 - **deps**: arquivos de entrada usados no comando
 - **outs**: arquivos gerados na saída
 - **name**: nome do estágio

### ⚙️ O que faz dvc stage add
Esse comando cria um novo stage no pipeline (adiciona no dvc.yaml) e também registra a dependência dos arquivos.

🧱 Sintaxe geral:

```bash
dvc stage add -n <nome_stage> -d <dependências> -o <saídas> <comando>
```

#### ✅ Exemplo real:
```bash
dvc stage add -n gerar_dados \
  -d etl/generator.py \
  -o data/raw/clientes_raw.csv \
  python scripts/run_etl.py --etapa gerar
```


Esse comando:
 - nomeia o stage como gerar_dados
 - declara que o script etl/generator.py é um arquivo de entrada (-d)
 - define que o arquivo de saída é data/raw/clientes_raw.csv (-o)
 - executa o comando python scripts/run_etl.py --etapa gerar

E o resultado disso será uma entrada como esta no dvc.yaml:

```yaml
stages:
  gerar_dados:
    cmd: python scripts/run_etl.py --etapa gerar
    deps:
      - etl/generator.py
    outs:
      - data/raw/clientes_raw.csv
```

### 🎯 Resumo dos principais comandos DVC

| Comando             | Função                                             |
| ------------------- | -------------------------------------------------- |
| `dvc init`          | Inicializa o DVC no projeto                        |
| `dvc stage add ...` | Adiciona um passo (stage) ao pipeline (`dvc.yaml`) |
| `dvc repro`         | Executa o pipeline conforme as dependências        |
| `dvc destroy`       | Remove o DVC do projeto                            |
| `dvc status`        | Verifica o que está desatualizado                  |
| `dvc dag`           | Visualiza graficamente a estrutura dos stages      |
