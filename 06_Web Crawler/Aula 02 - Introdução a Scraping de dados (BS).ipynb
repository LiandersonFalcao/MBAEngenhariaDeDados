## 2. BeatifulSoup

| Comando                                            | Descrição                                               | Exemplo                                           |
| -------------------------------------------------- | ------------------------------------------------------- | ------------------------------------------------- |
| `soup = BeautifulSoup(html, "html.parser")`        | Cria o objeto BeautifulSoup a partir do HTML.           | `soup = BeautifulSoup(resp.text, "html.parser")`  |
| `soup.find(tag, attrs={})`                         | Retorna o **primeiro elemento** que bate com o seletor. | `soup.find("a", {"class": "feed-post-link"})`     |
| `soup.find_all(tag, attrs={})`                     | Retorna **todos os elementos** que batem com o seletor. | `soup.find_all("a", {"class": "feed-post-link"})` |
| `soup.select(css_selector)`                        | Retorna lista de elementos usando **seletor CSS**.      | `soup.select("div.feed-post a.feed-post-link")`   |
| `soup.select_one(css_selector)`                    | Retorna apenas o **primeiro elemento** do seletor CSS.  | `soup.select_one("a.feed-post-link")`             |
| `element.get_text(strip=True)`                     | Extrai o **texto** limpo do elemento.                   | `title = a.get_text(strip=True)`                  |
| `element["atributo"]` ou `element.get("atributo")` | Extrai um **atributo HTML** (ex.: `href`, `src`).       | `link = a["href"]`                                |
| `soup.title.string`                                | Acessa o texto dentro da tag `<title>`.                 | `print(soup.title.string)`                        |
| `soup.prettify()`                                  | Retorna o HTML formatado de forma legível.              | `print(soup.prettify()[:500])`                    |
| `soup.text`                                        | Retorna **todo o texto cru** da página.                 | `print(soup.text[:500])`                          |
